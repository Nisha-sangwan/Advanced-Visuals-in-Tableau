**Question 1** : Create a Heatmap to analyze Treatment Cost by Department and Region.

**solution:**

1. Drag Department to the Rows shelf.

2. Drag Region to the Columns shelf.

3. Drag TreatmentCost to the Color card on the Marks shelf (ensure it aggregates as SUM or AVG based on preference).

4. Change the mark type dropdown on the Marks card from Automatic to Square.

5. Drag TreatmentCost to the Label card to display numeric values directly inside each cell.

**Question 2 :** Create a Highlight Table showing Satisfaction Score by Department.

**solution:**

1. Drag Department to the Rows shelf.

2. Double-click or drag SatisfactionScore to the Text card (ensure it is aggregated as AVG since individual scores range from 1 to 5).

3. Change the mark type dropdown to Square.

4. Hold Ctrl (or Cmd on Mac) and drag the AVG(SatisfactionScore) field from the Rows/Text card directly onto the Color card.

5. Edit the colors (e.g., Green-Gold-Red or Blue diverging) to instantly highlight high and low performing departments.

**Question 3** : Create a Bump Chart to show ranking of Departments based on Treatment Cost over months.

**solution:**

1. Drag AdmissionDate to the Columns shelf and change its detail level to Month (discrete/blue or continuous/green chronologically).

2. Drag TreatmentCost to the Rows shelf.Right-click SUM(TreatmentCost) in Rows, select Quick Table Calculation $\rightarrow$ Rank.Right-click the field again, select Compute Using $\rightarrow$ Department.

3.Drag Department to the Color card.

4. Reversed Axis Trick: Right-click the vertical axis, select Edit Axis, and check the Reversed box so that Rank 1 stays at the top.

5. Duplicate the Rows field (Ctrl + drag next to itself), turn it into a Dual Axis Chart, synchronize the axes, and change one mark type to Circle and the other to Line.

**Question 4** : Create a Butterfly Chart comparing Male vs Female patient count by Department.

**solution:**



1. Create a calculated field named Male Count: IF [Gender] = 'Male' THEN 1 ELSE 0 END.

2. Create a calculated field named Female Count: IF [Gender] = 'Female' THEN 1 ELSE 0 END.

3. Drag Department to the Rows shelf.

4. Drag SUM(Female Count) and SUM(Male Count) side-by-side onto the Columns shelf.

5. Double-click the horizontal axis for Female Count, and check the Reversed box under Scale options. This forces the Female bars to grow to the left, while Male bars grow to the right.




**Question 5** :  Create a Dual Axis Chart to compare Average Age and Treatment Cost by Department.

**solution**:

1. Drag Department to the Columns shelf.

2. Drag Age to the Rows shelf and change its aggregation to AVG(Age).

3. Drag TreatmentCost to the Rows shelf, placing it right next to AVG(Age).

4. Right-click the second pill (SUM(TreatmentCost)) on the Rows shelf and select Dual Axis.

5. On the Marks card, customize the marks for distinct visibility (e.g., change AVG(Age) to a Line chart and SUM(TreatmentCost) to a Bar chart).

**Question 6** : Create a Sunburst Chart showing Department → Diagnosis hierarchy by Treatment Cost.

**solution:**

1. Change the Mark type to Pie.

2. Drag Department to Color, and TreatmentCost to Angle.

3. Create a placeholder axis by typing MIN(1) directly into the Rows shelf twice.

4. On the second mark card layer, add both Department and Diagnosis to Color (or use Detail) and sort them hierarchically.

5. Make the charts a Dual Axis, synchronize, and adjust the size of the second pie chart to be slightly larger so it wraps around the inner core.




**Question 7** : Create a Dual Axis Chart comparing Treatment Cost and Satisfaction Score by Department.

**solution:**

1. Drag Department to the Columns shelf.

2. Drag TreatmentCost (SUM) to the Rows shelf.

3. Drag SatisfactionScore (AVG) to the Rows shelf right next to it.

4. Right-click the AVG(SatisfactionScore) pill and select Dual Axis.

5. Right-click the right axis and select Synchronize Axis if necessary (though because scales differ—millions vs. 1–5—leaving them unsynchronized but clearly labeled is best). Change one mark to a Bar and the other to a Gantt or Circle.

**Question 8 :** Create a Dashboard combining:

Heatmap

Dual Axis Chart

Sunburst Chart

Butterfly Chart

**solution:**

1. Click the New Dashboard icon at the bottom of the workbook.

2. Set the canvas size (e.g., Desktop Browser (1000 x 800)).

3. Drag a Dashboard Layout Container (Vertical or Horizontal) onto the canvas.

4. Drag and drop the Heatmap, Dual Axis Chart, Sunburst Chart, and Butterfly Chart worksheets into the grid.

5. Use the Use as Filter option (the funnel icon) on the Heatmap or Butterfly chart to make the entire dashboard dynamic and interactive!
